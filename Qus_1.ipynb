### Problem 1: Compute SVD manually for the matrix:

```python
print("Practice Problems:")
print("="*50)

# Problem 1: Manual SVD calculation
print("Problem 1: Compute SVD manually for the matrix:")
test_matrix = np.array([[1, 2], [3, 4], [5, 6]])
print(test_matrix)
```

```python

# Solution framework
print("\nSolution steps:")
print("1. Compute A^T * A")
AtA_test = test_matrix.T @ test_matrix
print(f"A^T * A = \n{AtA_test}")

print("2. Find eigenvalues and eigenvectors of A^T * A")
eigenvals_test, V_test = np.linalg.eigh(AtA_test)
print(f"Eigenvalues: {eigenvals_test}")
```
